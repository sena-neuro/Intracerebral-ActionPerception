```python
# Specificity Pkl Paths
pkl_path = "../Results/sig_spc/"
files = os.listdir(pkl_path)

list_of_dfs = []
for file in files:
    if file.endswith(".pkl"):
        df = pd.read_pickle(os.path.join(pkl_path,file))
        list_of_dfs.append(df)
specificity_df = pd.concat(list_of_dfs)
```

```python
with open (left_paint_file_path, "r") as left_paint_file:
    left_data=left_paint_file.read().splitlines()
    
with open (right_paint_file_path, "r") as right_paint_file:
    right_data=right_paint_file.read().splitlines()
```

```python
with open (rolla_paint_file_path, "r") as rolla_paint_file:
    rolla_paint_data=rolla_paint_file.read().splitlines()
```

```python
data = left_data + right_data
df_paint = pd.DataFrame(data,columns =["lead"])
```

```python
df_rolla_paint = pd.DataFrame(rolla_paint_data,columns =["lead"])
```

```python
for index, row in specificity_df.iterrows():
    row_name = row.name[1]
    subject_name = row.name[0]
    x = re.search("^[a-zA-Z]'?[1-9]$", row_name)
    if x: 
        row_name = row_name[0:1] + "_0" + row_name[1:]
        row_name = row_name.replace("'","")
    else:
        row_name = row_name.replace("'","_")
    lead_name = subject_name + "_" + row_name
    specificity = row.specificity
    found = df_rolla_paint["lead"].str.find(lead_name)
    place = found[found!=-1]
    if not len(place) == 0:
        row_no = place.index.values.astype(int)[0]
        col = place.values[0]
        df_rolla_paint.iloc[row_no] =df_rolla_paint.iloc[row_no].lead[:col]+ specificity + "_"+df_rolla_paint.iloc[row_no].lead[col:]
```

```python
f = open("df_rolla_paint.paint", "a")
f.write('\n'.join(df_rolla_paint.lead.values.tolist()))
f.close()
```
